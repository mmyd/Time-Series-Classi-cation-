```python
# Use 5-fold cross-validation to determine the best value of l.
import sklearn
from sklearn.cross_validation import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.feature_selection import RFE 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
cv_errors=[];Y_target=[];Y_test=[];num_fs=[];Y_score=[];best_fs=[];l_errors=[];
for l in range(1,21):
    fs=pd.DataFrame()
#break time series
    for instance in range(1,70):
        p=480//l
        ts=[]
        for q in range(l):
            idata=train_data[train_data["instance"]==instance][0+q*p:p+q*p]
            ts+=[idata['avg_rss12'].min(),idata['avg_rss12'].max(),idata['avg_rss12'].mean(),\
                idata['var_rss12'].min(),idata['var_rss12'].max(),idata['var_rss12'].mean(),\
                idata['avg_rss13'].min(),idata['avg_rss13'].max(),idata['avg_rss13'].mean(),\
                idata['var_rss13'].min(),idata['var_rss13'].max(),idata['var_rss13'].mean(),\
                idata['avg_rss23'].min(),idata['avg_rss23'].max(),idata['avg_rss23'].mean(),\
                idata['var_rss23'].min(),idata['var_rss23'].max(),idata['var_rss23'].mean()]
        fs=pd.concat([fs,pd.DataFrame([ts])])
    train_target=np.append(np.array([1]*9),np.array([0]*60))
# Use 5-fold cross-validation 
    skf = StratifiedKFold(train_target, n_folds=5)
    errors=[];
    # keep the train data in fs1 when L=1
    if l==1:
        fs1=fs
    for train_index, test_index in skf:
        features=[];
        fs_train, fs_test = fs.values[train_index], fs.values[test_index]
        target_train, target_test = train_target[train_index], train_target[test_index]
        rfecv =   RFECV(estimator=LogisticRegression(),cv=5)
        rfecv.fit(fs_train,target_train)
        test_pred=rfecv.predict(pd.DataFrame(fs_test))
        errors.append(sum(target_test!=test_pred))
        # keep the selected features, test_errors for L=1 when doing 5-folds CV.
        if l==1:
            best_fs.append(rfecv.support_)
            l_errors=errors;
            Y_test.append(test_pred);Y_target.append(target_test);
            Y_score.append(rfecv.predict_proba(pd.DataFrame(fs_test))[:,1])
    cv_errors.append(sum(errors)/len(errors))
plt.title("Plot of CV test errors") 
plt.xlabel("L")
plt.ylabel("error rate")
plt.plot(list(range(1,21)),cv_errors,"b-o")
plt.show()
```

```python
#the best value of L is 1
#Report the confusion matrix and show the ROC and AUC for your classiﬁer on train data. 
from sklearn.metrics import confusion_matrix,roc_auc_score,roc_curve,auc

y_true=np.concatenate((Y_target[0],Y_target[1],Y_target[2],Y_target[3],Y_target[4]))
y_pred=np.concatenate((Y_test[0],Y_test[1],Y_test[2],Y_test[3],Y_test[4]))
y_score=np.concatenate((Y_score[0],Y_score[1],Y_score[2],Y_score[3],Y_score[4]))
print('confusion matrix:\n',confusion_matrix(y_true, y_pred))
fpr, tpr, thresholds = roc_curve(y_true, y_score)
plt.figure()
#plt.plot(fpr[1], tpr[1], color='darkorange',label='ROC curve (area = %0.2f)' %  roc_auc_score(y_true, Y_score))
plt.plot(fpr,tpr,"b-o")
plt.plot([0, 1], [0, 1], color='red', linestyle='--',label='ROC curve (area = %0.5f)' %  roc_auc_score(y_true, y_score))
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

```python
# Multi-class Classiﬁcation (The Realistic Case) 
import warnings
from sklearn import preprocessing
from sklearn.linear_model import LassoCV,LogisticRegressionCV
cv_errors3=[];
#warnings.filterwarnings('ignore')
for l in range(1,21):
    multi_fs=pd.DataFrame()
#break time series
    for instance in range(1,70):
        p=480//l
        ts=[]
        for q in range(l):
            idata=train_data[train_data["instance"]==instance][0+q*p:p+q*p]
            ts+=[idata['avg_rss12'].min(),idata['avg_rss12'].max(),idata['avg_rss12'].mean(),\
                idata['var_rss12'].min(),idata['var_rss12'].max(),idata['var_rss12'].mean(),\
                idata['avg_rss13'].min(),idata['avg_rss13'].max(),idata['avg_rss13'].mean(),\
                idata['var_rss13'].min(),idata['var_rss13'].max(),idata['var_rss13'].mean(),\
                idata['avg_rss23'].min(),idata['avg_rss23'].max(),idata['avg_rss23'].mean(),\
                idata['var_rss23'].min(),idata['var_rss23'].max(),idata['var_rss23'].mean()]
        multi_fs=pd.concat([multi_fs,pd.DataFrame([ts])])
    multi_normalized= preprocessing.normalize(multi_fs, norm='l2')
    multi_normalized= pd.DataFrame(multi_normalized)
    class1=np.append(np.array(['bending1']*5),np.array(['bending2']*4))
    class2=np.append(class1,np.array(['cycling']*12))
    class3=np.append(class2,np.array(['lying']*12))
    class4=np.append(class3,np.array(['sitting']*12))
    class5=np.append(class4,np.array(['standing']*12))
    class6=np.append(class5,np.array(['walking']*12))
# Use 5-fold cross-validation 
    skf = StratifiedKFold(class6, n_folds=5)
    errors=[];
    for train_index, test_index in skf:
        fs_train, fs_test = multi_normalized.values[train_index], multi_normalized.values[test_index]
        target_train, target_test = class6[train_index], class6[test_index] 
        multi = LogisticRegressionCV(penalty='l1', solver='saga',multi_class='multinomial',cv=5,max_iter=2).fit(fs_train, target_train)
        errors.append(1-multi.score(fs_test, target_test))
    cv_errors3.append(sum(errors)/len(errors))
plt.title("Plot of CV errors") 
plt.xlabel("L")
plt.ylabel("error")
plt.plot(list(range(1,21)),cv_errors3,"b-o")
plt.show()
```
